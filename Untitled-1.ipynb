```python
import matplotlib.pyplot as plt
import numpy as np

# Generating practical data for a simple pendulum experiment
L = 1  # Length of pendulum in meters
g = 9.81  # Acceleration due to gravity in m/s^2
T_theoretical = 2 * np.pi * np.sqrt(L / g)

# Generating some practical data: Period measurements for different trials
np.random.seed(42)  # For reproducibility
num_trials = 10
periods = T_theoretical + np.random.normal(0, 0.05, num_trials)  # Adding noise

# Trial numbers
trials = np.arange(1, num_trials + 1)

# Plotting the experimental period vs trial number
plt.figure(figsize=(10, 6))
plt.plot(trials, periods, marker='o', color='g', linestyle='-', label='Measured Period')
plt.axhline(y=T_theoretical, color='r', linestyle='--', label='Theoretical Period')
plt.title('Simple Pendulum Experiment - Period vs. Trial Number')
plt.xlabel('Trial Number')
plt.ylabel('Period (s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating practical data for a simple pendulum experiment to show Length vs Period (T)
# Assuming gravitational acceleration g = 9.81 m/s^2

# Different lengths of the pendulum
lengths = np.linspace(0.1, 2, 10)  # Lengths from 0.1 m to 2 m (10 points)

# Calculating the period for each length using the formula:
# T = 2 * pi * sqrt(L / g)
g = 9.81  # Acceleration due to gravity in m/s^2
periods = 2 * np.pi * np.sqrt(lengths / g)

# Plotting Length vs Period
plt.figure(figsize=(10, 6))
plt.plot(lengths, periods, marker='o', color='b', linestyle='-', label='Period (T)')
plt.title('Simple Pendulum Experiment - Length vs. Period')
plt.xlabel('Length (m)')
plt.ylabel('Period (s)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for mean and standard deviation of each metric
metrics = ["Speech Recognition Accuracy (%)", "Task Execution Time (seconds)", "Misclassification Rate (%)", "Task Success Rate (%)"]
means = [94.0, 44.8, 8.2, 85.4]
std_devs = [1.5, 3.2, 3.8, 4.6]

# Generate x values for probability distributions
x_values = {
    "Speech Recognition Accuracy (%)": np.linspace(90, 100, 200),
    "Task Execution Time (seconds)": np.linspace(35, 55, 200),
    "Misclassification Rate (%)": np.linspace(0, 20, 200),
    "Task Success Rate (%)": np.linspace(60, 100, 200)
}

# Generate probability distributions for each metric
distributions = {
    metric: (x_values[metric], (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values[metric] - mean) / std_dev) ** 2))
    for metric, mean, std_dev in zip(metrics, means, std_devs)
}

# Plotting the probability distributions
plt.figure(figsize=(12, 8))

for i, (metric, (x, y)) in enumerate(distributions.items(), start=1):
    plt.subplot(2, 2, i)
    plt.plot(x, y, label=metric)
    plt.fill_between(x, y, alpha=0.3)
    plt.title(f"Probability Distribution of {metric}")
    plt.xlabel(metric)
    plt.ylabel("Probability Density")
    plt.legend()

plt.tight_layout()
plt.show()
```
